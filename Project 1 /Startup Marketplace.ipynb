# Задача 1: Оценить месячный retention в оформление заказа с помощью когортного анализа.

В рамках исследования необходимо:

Исследовать датасет и определить, какой вид заказа будет учитываться в retention\
Построить месячный retention\
Проанализировать, чему равен медианный retention 1-го месяца\
Найти когорту с самым высоким retention на 3-й месяц.

```python
# Посмотрим на количество заказов по статусам.
orders['order_status'].value_counts()
```

Для расчета retention будем использовать только доставленные товары, так как у них пройден весь цикл от создания заказа до получения клиентом.

```python
# Мерджим первые две таблицы, и делаем копию для дальнейшего анализа.
result = orders.merge(customers, how='left', on='customer_id')
result_copy = result.copy()
# Отбираем заказы со статусом "доставлен"
result_copy = result_copy.query('order_status=="delivered"')
```

```python
# Извлекаем месяц и день заказа
result_copy['order_period'] = result_copy['order_purchase_timestamp'].dt.to_period('M')
result_copy['order_period_day'] = result_copy['order_purchase_timestamp'].dt.to_period('D')
```

```python
# Посмотрим на распределение заказов по месяцам
result_copy.groupby('order_period', as_index=False).size()
```

За сентябрь и декабрь 2016 года в базе имеется только по одному заказу, в октябре тоже очень небольшое количество заказов относительно остальных месяцев, а в ноябре данные отсутсвуют, поэтому для кагортного анализа возьмем данные с начала 2017 года:

```python
result_copy = result_copy.query('order_purchase_timestamp>="01-01-2017"')
```

```python
# Присваиваем когорты на основе даты первого действия для каждого пользователя
result_copy['cohort'] = result_copy.groupby('customer_unique_id')['order_purchase_timestamp']\
                                   .transform('min').dt.to_period('M')
result_copy['cohort_day'] = result_copy.groupby('customer_unique_id')['order_purchase_timestamp']\
                                       .transform('min').dt.to_period('D')
```

```python
# Рассчитываем номер периода в днях (месячные периоды, деля на 30)
result_copy['period_number_month'] = np.floor((result_copy.order_period_day - result_copy.cohort_day)\
                                       .apply(attrgetter('n')) / 30)
```

```python
# Агрегируем данные по когортам и рассчитанному номеру периода (в месяцах)
df_cohort = result_copy.groupby(['cohort', 'period_number_month'])\
                       .agg(n_customers=('customer_unique_id', 'nunique'))\
                       .reset_index()
df_cohort['period_number_month'] = df_cohort['period_number_month'].astype(int)
df_cohort.head()
```

```python
# Создаем сводную таблицу для когортного анализа
cohort_pivot = df_cohort.pivot_table(index='cohort', columns='period_number_month', values='n_customers')
cohort_pivot
```

```python
# Рассчитываем размеры когорт (первый столбец сводной таблицы)
cohort_size = cohort_pivot.iloc[:, 0]
# Вычисляем коэффициенты удержания, деля на размер когорты
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
retention_matrix
```

```python
# Строим тепловую карту когортного анализа
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(16, 12), sharey=True, gridspec_kw={'width_ratios': [1, 11]})

    # Тепловая карта для коэффициентов удержания
    sns.heatmap(retention_matrix,
                mask=retention_matrix.isnull(),
                annot=True,
                fmt='.1%',
                cmap='RdYlGn',
                ax=ax[1])
    ax[1].set_title('Monthly Retention', fontsize=16)
    ax[1].set(xlabel='№ периода', ylabel='Когорта')

    # Тепловая карта для размеров когорт
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df,
                annot=True,
                cbar=False,
                fmt='g',
                cmap=white_cmap,
                ax=ax[0])

    fig.tight_layout()
    plt.show()
```

Мы построили тепловую карту когортного анализа и видим, что возвращаемость клиентов в маркетплейсе очень мала. Отсюда можем сделать вывод, что в компании имеется какая-то проблема, сталкиваясь с которой, клиенты полчают негативный опыт и не хотят совершать повторный заказ. 

```python
# Находим медианный retention 1-го месяца и когорту с самым высоким retention на 3-й месяц
mediana = retention_matrix[1].median()
print(f'Медианный retention 1-го месяца равен {mediana.round(4)}.')
```

```python
cohort_m = retention_matrix[retention_matrix[3] == retention_matrix[3].max()].index.tolist()
max_ret = retention_matrix[3].max()
print(f'Когортой с самым высоким retention на 3-й месяц является {str(cohort_m)[9:16]}, retension равен {max_ret.round(4)}.')
```

# Задача 4. Выбрать одну из 3 основных гипотез с помощью фреймворка ICE.

Для расчёта Impact возьмем данные с июня 2017 года. Конверсию в повторный заказ берем равной величине медианного retention 1-го месяца.\
Значения уверенности в гипотезах и сложности их реализации нам дала команда:

```python
confidence1 = 8
ease1 = 6
confidence2 = 10
ease2 = 4
confidence3 = 5
ease3 = 9
```

```python
# Напишем функцию для перевода импакта в шкалу от 1 до 10
def impact(x):
    ranges = [(0, 50, 1),
             (51, 150, 2),
             (151, 350, 3),
             (351, 750, 4),
             (751, 1550, 5),
             (1551, 3150, 6),
             (3151, 6350, 7),
             (6351, 12750, 8),
             (12751, 25550, 9),
             (25551, 51150, 10)] 
    for lower, upper, impact in ranges:
        if lower <= x <= upper:
            return impact
```

# Гипотеза 1. Если исправим баг в системе процессинга заказов, то клиентам не придётся сталкиваться с проблемой отмены заказа, вследствие чего количество доставленных заказов увеличится. Считаем, что мы таким образом избавимся от всех отмен.

```python
# Посчитаем количество отмененных заказов
result['order_period'] = result['order_purchase_timestamp'].dt.to_period('M')
from_june = result.query('order_period >= "2017-06"')
canceled = from_june.query('order_status == "canceled"')['order_id'].count()
```

```python
ice1 = impact(canceled) * confidence1 * ease1 
print(f'Имакт первой гипотезы равен {impact(canceled)}, ICE {ice1}')
```

# Гипотеза 2. Если сократим время до отгрузки заказа, то клиенты перестанут получать свой заказ с запаздыванием, вследствие чего количество заказов увеличится за счёт повторных заказов.

```python
# Отбираем доставленные заказы и считаем сколько дней прошло от подтверждения оплаты до отгрузки
from_june_deliver = from_june.query('order_status == "delivered"')
from_june_deliver['dispatch_time'] = (from_june_deliver['order_delivered_carrier_date']\
                                      - from_june_deliver['order_approved_at']).dt.days
from_june_deliver['dispatch_time'].value_counts().head(10)
```

Эталоном отгрузки будем считать до 4 дней, так как заказ подтвержденный в пятницу могут отгрузить в понедельник.

```python
# Разделим датафрейм на группы с долгой и быстрой отгрузкой
fast_dispath = from_june_deliver.query('dispatch_time <= 3') # первый день находится на нуле
slow_dispath = from_june_deliver.query('dispatch_time > 3')
```

```python
# Посчитаем сколько человек из группы с быстрой отгрузкой совершили 2 и более заказов
fast_dispath_total = fast_dispath.groupby('customer_unique_id').nunique()['order_id'].count()
fast_dispath_two_orders = fast_dispath.groupby('customer_unique_id').nunique().query('order_id >= 2').shape[0]
# Считаем конверсию во второй заказ в быстрой группе
conv2 = fast_dispath_two_orders/fast_dispath_total
```

```python
# Получим число пользователей, у которых была долгая отгрузка, по ним мы будем расчитывать импакт
slow_dispath_total = slow_dispath.groupby('customer_unique_id').nunique()['order_id'].count()
share_slow_dispath = slow_dispath_total*100/(from_june_deliver.groupby('customer_unique_id').nunique()['order_id'].count())
print(f'Доля людей с долгой отгрузкой составляет {round(share_slow_dispath)}%')
```

При расчете импакта берем конверсию во второй заказ, полученную в группе с быстрой отгрузкой и умножаем на группу с долгой.

```python
impact2 = slow_dispath_total*conv2 
```

```python
ice2 = impact(impact2) * confidence2 * ease2 
print(f'Импакт второй гипотезы равен {impact(canceled)}, ICE {ice2}')
```
